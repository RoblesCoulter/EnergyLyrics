```python
import json
import pandas as pd
from pandas import DataFrame as df
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
df_word2vec = pd.read_pickle('Dataset/df_candidate_vec.pkl')
```

```python
moedict = json.load(open('Dataset/dict-revised.json', 'r'))
```

```python
syn_cluster = dict([(word, None) for word in df_word2vec.index])
```

```python
moedict
```

```python
cluster_id = 0
for item in moedict:
    if item['title'] not in df_word2vec.index: continue
    try:
        item['heteronyms'][0]['definitions']
    except:
        continue
    
    syns = []
    for defs in item['heteronyms'][0]['definitions']:
        if 'synonyms' not in defs.keys(): continue
        if len(defs['synonyms'].split('﹑')) > len(defs['synonyms'].split(',')):
            syns += defs['synonyms'].split('﹑')
        else:
            syns += defs['synonyms'].split(',')
    
    flag = 0
    for syn in syns:
        if syn not in df_word2vec.index: continue
        syn_cluster[syn] = cluster_id
        flag = 1
    
    if flag:
        syn_cluster[item['title']] = cluster_id
        cluster_id += 1
```

```python
df([x[1]!=None for x in syn_cluster.items()]).sum()
```

```python
df_syn_olp_cluster = df([int(syn_cluster[w]) if syn_cluster[w] != None else None for w in df_word2vec.index], index=df_word2vec.index)
```

```python
df_syn_olp_cluster.to_pickle('df_syn_olp_cluster_chn.pkl')
```
